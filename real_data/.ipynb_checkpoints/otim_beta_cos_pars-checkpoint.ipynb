### Implementação do modelo

Condições iniciais: 

```python
data_s = data.loc[data.data_iniSE < start_date][['casos_41010', 'casos_41009', 'casos_41020']]
Ii0 = data_f.iloc[0, 1:].values

# dados para a minimização
samples =  data_f.iloc[:, 1:].values

samples.shape
```

Funções: 

```python
def comp_epsilon(theta, alpha, dist_table_path):
    """
    Compute the epsilon matrix for the specified parameters.

    Parameters:
        theta (float): Scaling parameter.
        alpha (float): Exponent for distance decay.
        dist_table_path (str): Path to the CSV file containing the distance matrix.

    Returns:
        numpy.ndarray: Epsilon matrix.
    """
    # Ordem das regiões:
    # Cascavel, Foz do Iguaçu, Francisco Beltrao, Pato Branco, Toledo

    # Densidade populacional
    dens_pop = [49.42, 109.37, 52.28]; 

    # Carregar matriz de distâncias
    D = pd.read_csv(dist_table_path)
    D = D.iloc[:, 1:4].to_numpy()  # Select the relevant columns and convert to NumPy array

    # Inicializar matriz epsilon
    n, m = 3, 3
    epsilon = np.zeros((n, m))

    # Calcular elementos da matriz epsilon
    for i in range(n):  # Loop sobre as linhas
        for j in range(m):  # Loop sobre as colunas
            H_i = dens_pop[i]
            H_j = dens_pop[j]

            epsilon[i, j] = theta * H_i * H_j / (D[i, j] ** alpha)

    # Substituir valores infinitos na diagonal principal
    epsilon[np.isinf(epsilon)] = 0

    return epsilon

def beta_cos(temp, beta_0, eta, zita, phi):
    """
    Compute the beta vector based on the temperature and given parameters.

    Parameters:
        temp (numpy.ndarray): Array of temperature values.
        beta_0 (list or numpy.ndarray): Base beta values for each region (length 5).
        eta (list or numpy.ndarray): Amplitude values for each region (length 5).
        zita (list or numpy.ndarray): Frequency values for each region (length 5).
        phi (list or numpy.ndarray): Phase shift values for each region (length 5).

    Returns:
        numpy.ndarray: Beta matrix of shape (len(temp), 5).
    """
    # Initialize the beta matrix with NaN values
    beta_vec = np.full((len(temp), 3), np.nan)

    # Compute beta values for each region
    for i in range(3):
        beta_vec[:, i] = beta_0[i]*(1 + eta[i] * np.cos(zita[i] * (temp + phi[i])))

    beta_vec[beta_vec < 0] = 0
    
    #beta_vec[beta_vec > 1] = 1

    return beta_vec

def model_meta_pop_aux_no_R(gamma, tau, capa, beta, epsilon, N, S_K, E_K, I_K):
    """
    Update S, E, and I for the next time step in a meta-population model without R.

    Parameters:
        gamma (float): Recovery rate.
        tau (float): Transition rate from E to I.
        capa (float): Emigration rate.
        beta (float): Transmission rate.
        epsilon (numpy.ndarray): Interaction matrix.
        N (numpy.ndarray): Total population in each region.
        S_K (numpy.ndarray): Susceptible population in each region.
        E_K (numpy.ndarray): Exposed population in each region.
        I_K (numpy.ndarray): Infectious population in each region.

    Returns:
        tuple: Updated S_K, E_K, and I_K arrays.
    """
    # Compute lambda_i_K
    lambda_i_K = beta * (I_K / N)

    # Update S, E, and I
    S_K_1 = S_K + capa * N - lambda_i_K * S_K - lambda_ij(beta, epsilon, I_K, N) * S_K - capa * S_K
    E_K_1 = E_K + lambda_i_K * S_K + lambda_ij(beta, epsilon, I_K, N) * S_K - tau * E_K - capa * E_K
    I_K_1 = I_K + tau * E_K - gamma * I_K - capa * I_K

    return S_K_1, E_K_1, I_K_1

def lambda_ij(beta, epsilon, I, N):
    """
    Compute the lambda_ij matrix and the resulting lambda vector.

    Parameters:
        beta (numpy.ndarray): Array of beta values (length 5).
        epsilon (numpy.ndarray): Interaction matrix (5x5).
        I (numpy.ndarray): Array of infectious populations (length 5).
        N (numpy.ndarray): Array of total populations (length 5).

    Returns:
        numpy.ndarray: Lambda vector (length 5).
    """
    # Initialize the lambda matrix with NaN
    lmbda = np.full((3, 3), np.nan)

    # Compute each element of the lambda matrix
    for i in range(3):
        for j in range(3):
            lmbda[i, j] = epsilon[i, j] * beta[i] * I[j] / N[j]

    # Sum over columns and transpose the result
    lmbda_end = np.sum(lmbda, axis=1)

    return lmbda_end
```

```python
def apply_model(beta, epsilon, Ii0, data_s): 
    tfinal = len(beta)

    Si = np.full((tfinal, 3), np.nan)
    Ei = np.full((tfinal, 3), np.nan) 
    Ii = np.full((tfinal, 3), np.nan)
    Ri = np.full((tfinal, 3), np.nan)
    
    # condições iniciais 
    Ei_K = Ii0
    Ii_K = Ii0 
    Ri_K = data_s.sum().values
    Si_K = Ni - Ei_K - Ii_K - Ri_K
    
    Si[0,:] = Si_K
    Ei[0,:] = Ei_K
    Ii[0,:] = Ii_K
    for a in np.arange(1, tfinal):

        [S_K_1,E_K_1,I_K_1] = model_meta_pop_aux_no_R(gamma, tau, capa, beta[a-1, :], epsilon, Ni, 
                                                        Si[a-1, :], Ei[a-1,:], Ii[a-1, :])  
                 
        Si[a,:] = S_K_1
        Ei[a,:] = E_K_1
        Ii[a,:] = I_K_1

    return Si, Ei, Ii
            
```

### Otimizando os parâmetros do beta 

```python
def obj_fun(params, samples, Ii0, data_s, temp):
    
    pars = params.valuesdict()
    beta_1 = pars["beta_1"]
    beta_2 = pars["beta_2"]
    beta_3 = pars["beta_3"]
      
    eta_1 = pars["eta_1"]
    eta_2 = pars["eta_2"]
    eta_3 = pars["eta_3"]

    zita_ = pars["zita"]
    phi_ = pars["phi"]
    theta = pars['theta']
    
    epsilon = comp_epsilon(theta, alpha, dist_table_path = '../data/dist_table_model.csv')

    beta_0 = [beta_1, beta_2, beta_3] 
    eta = [eta_1,eta_2,eta_3]
    zita = [zita_, zita_, zita_]
    phi = [phi_, phi_, phi_]
    
    beta = beta_cos(temp, beta_0, eta, zita, phi)
    
    Si, Ei, Ri = apply_model(beta, epsilon, Ii0, data_s)

    soma = 0
    for i in np.arange(3):
        soma = soma + root_mean_squared_error(samples[:,i], Ei[:, i])

    return soma 
```

```python
params = Parameters()
params.add("beta_1", value=1, min=0.1, max=2)
params.add("beta_2", value=1, min=0.1, max=2)
params.add("beta_3", value=1, min=0.1, max=2)

params.add("eta_1", value=0.5, min=0.1, max=1)
params.add("eta_2", value=0.5, min=0.1, max=1)
params.add("eta_3", value=0.5, min=0.1, max=1)

params.add("zita", value=0.1, min=0.01, max=0.3, vary = True)
params.add("phi", value = -7, min=-30, max=0)
params.add("theta", value=1.2, min=0.9, max=2.5, vary = False)

out = lm.minimize(obj_fun, params, args = ( samples, Ii0, data_s, np.arange(0, data_f.shape[0])), method="diferential_evolution")

pars = out.params
pars = pars.valuesdict()
```

```python
def plot_preds(data_f, Ei, title = ''): 

    fig,ax = plt.subplots(1,3, figsize = (15, 3.5))

    ax_ = ax.ravel()
    
    for ax1, col, idx in zip(ax_, ['casos_41010', 'casos_41009', 'casos_41020'],
                            [0,1,2,3,4]):
        
        ax1.plot(data_f.data_iniSE, data_f[col] , marker ='.', ls = '', color = 'black')
        ax1.plot(data_f.data_iniSE, Ei[:, idx])
        ax1.set_title(f'{col[-5:]} - RMSE: {round(root_mean_squared_error(data_f[col], Ei[:, idx]), 2)}')
    
    plt.suptitle(title)
    plt.show()
    
```
